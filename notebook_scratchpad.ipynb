```python
def mod_inverses(m, n, m_0=1, m_1=0, n_0=0, n_1=1):

    print(f"{m:5} {n:5} {m % n:5} {m_0:12} {m_1:6} {n_0:10} {n_1:6}")

    if (m % n) == 0:
        return (m_1, n_1, n)
    else:
        m_2 = m_0 - m_1 * (m // n)
        n_2 = n_0 - n_1 * (m // n)
        m, n = n, m % n
        return mod_inverses(m, n, m_1, m_2, n_1, n_2)


print()
p = 2348
q = 907

p_n, q_n, gcd = mod_inverses(p, q)

print(f"\n{gcd:11} {p_n:25} {q_n:17}")
print(f"{gcd:11} {p_n % p:25} {q_n % p:17} {'(modulo ' + str(p) + ')':>16}\n")

print(
    f"\nSince {p_n} and {q_n} are relatively prime and the "
    f"linear combination\n{p_n % p}({p}) + {q_n % p}({q}) = {p * p_n + q * q_n}"
    f" (mod {p}), the GCD of {p} and {q} is {gcd}\n"
)
```
